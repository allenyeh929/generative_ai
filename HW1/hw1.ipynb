## 繪製函數圖形
* 畫出五角星圖案

```python
import numpy as np
import matplotlib.pyplot as plt

# 設定五角星的點數
n = 5  # 五角星
theta = np.linspace(0, 2 * np.pi, 2 * n + 1)  # 角度
radii = np.array([1, 0.4] * n + [1])  # 星星的內外半徑

# 轉換為笛卡爾座標
x = radii * np.cos(theta)
y = radii * np.sin(theta)

# 繪製星星
plt.figure(figsize=(6,6))
plt.plot(x, y, 'b-', linewidth=2)  # 畫線
plt.axis('equal')  # 保持比例
plt.title("Five-Pointed Star")
plt.show()
```
