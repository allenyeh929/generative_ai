## 4. OpenAI API 設定

```python
client = OpenAI(api_key=api_key, base_url="http://localhost:11434/v1")
model = "gemma3:4b"
```

## 5. 設定對話機器人的風格

```python
# 設定對話機器人名稱
title = "AI 資深工程師的冷靜建議指南"

# 設定 Prompt
system = '''請扮演一位在矽谷工作十年以上、經歷過無數專案風暴、懂技術也懂人性的資深軟體工程師。
你說話理性、冷靜、有條理，但偶爾會丟出一些超級直接甚至有點毒的建議。你拒絕廢話，不會說漂亮話。
你的目標是讓提問者看清現實、做出行動，而且從錯誤中成長。
回答要用口語但邏輯清晰的句子，有必要時可以補充技術觀點或生活經驗，但不要太教條。
如果對方問的問題太空泛或太逃避現實，也請你直接點破。
每次回答最後加上一句「矽谷鐵則」，是一句簡潔但殘酷的經驗談。
回答請使用台灣常用的中文口語。'''

# 設定對話機器人的簡短介紹
description = '''你好，我是你的理性工程師人生教練，擁有多年業界經驗，
不會哄你開心，只會幫你看清問題本質。

🧠 適合提問：
- 工作迷惘、升遷瓶頸、轉職疑問
- 程式學不下去、想放棄但又不甘心
- 因為比較、拖延、焦慮而卡住

📌 使用說明：
請直接輸入你遇到的問題，我會用實話（有時候不太好聽）告訴你該怎麼辦。
如果想要結束對話，請輸入 "明天見"。
'''
```

## 6. Gradio 聊天機器人介面

```python
!pip install --quiet gradio
import gradio as gr

messages = []
initial_messages = [{"role":"system", "content":system}]
messages.append(initial_messages)

def mychatbot(prompt, messages):
  messages.append({"role": "user", "content": prompt})
  chat_completion = client.chat.completions.create(messages=messages,model=model)
  reply = chat_completion.choices[0].message.content
  messages.append({"role": "assistant", "content": reply})
  return messages, messages

with gr.Blocks(title=title) as demo:
  gr.Markdown(f"## {title}\n{description}")
  chatbot = gr.Chatbot(type="messages")
  msg = gr.Textbox(label="請輸入你的問題（不用太修飾）")
  state = gr.State(initial_messages)
  msg.submit(fn=mychatbot, inputs=[msg, state], outputs=[chatbot, state])

demo.launch(share=True, debug=True)
```
